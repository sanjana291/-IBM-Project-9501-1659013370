```python
# Splitting data as independent and dependent
# Removing index column in independent dataset
x=df.iloc[:,1:31]
y=df.Result
```

```python
#Scale the independent variables
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
xscale=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
#  Split the data into training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xscale,y,test_size=.2,random_state=10)
```

```python
# model building
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
```

```python
model1.fit(xtrain,ytrain)
```

```python
trainpred1=model1.predict(xtrain)
testpred1=model1.predict(xtest)
```

```python
# evaluation metics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
pd.crosstab(ytest,testpred1)
```

```python
accuracy_score(ytest,testpred1)
```

```python
accuracy_score(ytrain,trainpred1)
```

```python
print(classification_report(ytest,testpred1))
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier()
```

```python
model2.fit(xtrain,ytrain)
```

```python
trainpred2=model2.predict(xtrain)
testpred2=model2.predict(xtest)
```

```python
print(confusion_matrix(ytest,testpred2))
```

```python
print(accuracy_score(ytest,testpred2))
print(accuracy_score(ytrain,trainpred2))
```

```python
print(classification_report(ytest,testpred2))
```
