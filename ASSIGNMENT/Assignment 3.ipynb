### 7. Check for Categorical columns and perform encoding.

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df.Sex=le.fit_transform(df.Sex)
df.head()
```

### 8. Split the data into dependent and independent variables.

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.Age
y.head()
```

### 9. Scale the independent variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
xscaler=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
xscaler.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures()
```

```python
xp=pr.fit_transform(xscaler)
xp
```

### 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xp,y,test_size=0.2,random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

### 11. Build the Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

### 12. Train the Model

```python
model.fit(x_train,y_train)
```

```python
model_pred=model.predict(x_train)
```

```python
from sklearn import metrics
```

```python
#R2-score
print(metrics.r2_score(y_train,model_pred))
```

### 13. Test the Model

```python
test_model_pred = model.predict(x_test)
```

```python
#R2-score
print(metrics.r2_score(y_test,test_model_pred))
```
